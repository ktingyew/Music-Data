# Reprocessing Gender

When it comes to analysing on gender lines, I will use 2 approaches:

1. First is to use the existing 4 labels under the original 'Gender' tag.
2. Second is to use the boolean values under the newly created 'M' and 'F' columns'

For the latter, we split the variable `Gender` into two binary variables: `M` and `F`. We do this by making 2 new columns:

|Gender| M |F |
|---|---|---|
Male|1|0|
Female|0|1|
Both|1|1|
N.A.|0|0

This means that for songs labelled 'Both' in the Gender tag, they will be counted as distinct songs when I filter purely under variables `M` or `F` in isolation. This is because such songs will have a value of `1` under both columns (double counting).

That said, I will switch between the 2 approaches from time to time. I recognise that 'Both' songs are actually quite populous.

````python
GLY['M'] = GLY['Gender'].map(lambda val : 1 if (val == 'Male' or val == 'Both') else 0 )
GLY['F'] = GLY['Gender'].map(lambda val : 1 if (val == 'Female' or val == 'Both') else 0 )

KRAG['M'] = KRAG['Gender'].map(lambda val : 1 if (val == 'Male' or val == 'Both') else 0 )
KRAG['F'] = KRAG['Gender'].map(lambda val : 1 if (val == 'Female' or val == 'Both') else 0 )
````

````python
plot = GLY[['M','F']].agg('sum').plot(kind='bar', figsize = (4, 5), color = 'purple')
plot.set_title("Breakdown of Songs by Gender (M/F)")
for i in plot.patches:
    plot.text(i.get_x()+0.175, i.get_height(), i.get_height(), fontsize = 15)
````

# Breakdown By Rating

All songs are given a rating - a number of stars, up to 5, with half-stars allowed. This allows for 11 distinct values (0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0).

Personally, I don't consider 0.0 stars as a valid rating. To me, a song with 0 stars means that it is unrated, and thus is awaiting rating. As such, songs that I rate the most poorly are assigned 0.5 stars instead. Consequently, with 10 remaining rating values, the median value equates to 2.75 star, what I now refer to as the 'median rating'.

## Average 

Let's compute the average rating across my entire library.

````python
def str_round(val, dp):
    return str(round(val, dp))

avg_rating = KRAG['Rating'].agg('mean')
avg_discrepancy = avg_rating - 2.75
total_discrepancy = avg_discrepancy * KRAG.shape[0]

print("The average rating is: " + str_round(avg_rating, 4))
print("Given median rating should be 2.75, this means that your current library average rating is off by: " + 
      str_round(avg_discrepancy, 5) )
print("Library has " + str(KRAG.shape[0]) + 
      " songs. Therefore, to keep in line with median rating, you need to add/remove ", end="" )
print( str_round(total_discrepancy, 2) + " stars.")
````

## Discussion about Median Rating, and the desire for Balance

Understandably, a question pops up: Why should average rating be the same as the median rating of 2.75? I don't really have any convincing argument, other than my desire to minimise the skewing of songs' ratings (as an aggregate) towards either "likeable" or "unlikeable" sides of the spectrum. And as such, I would like this to happen. So (as of 21 April 2020), I am making a gradual effort in shifting the balance to the median. To do so, I will be re-evaluating (a highly subjective process) what certain (number of) stars mean for me when it comes to my perceived enjoyment of a song. And in this regard, I will be adopting a "harsher" stance - i.e. less generous with giving away stars (because the current situation is that the average rating is above the median rating). Note that as aforementioned, this is a gradual process. On top of that, I am most definitely not overhauling the rating assignments of my entire music library - that's way too cumbersome. What I'll be doing is that I will evaluate a song as it pops up in 'Now Playing'. Of course, this is not ideal as such a methodology means that I will miss some songs that "deserves" to have their rating dropped by the time I finally have my average rating hit the median rating.

I will be creating a list to monitor. Of course this could have been "codified", but that's not worth my time. This reshifting process that I am undertaking is a once-off event, and once I "fixed it", there's no point in constantly measuring the average rating, save for every now and then, when the balance gets skewed way too much for my liking, and it calls for another rebalancing.

### Discussion

matplotlib is less preferred because unlike Seaborn, it does not have an easy way to colours the bars distinctly.

When it comes to `barplot` vs `countplot`, using `barplot` is worst because it is not designed to do something simple like counting instances! You can see how we have to input a dataframe into 2 of its parameters. The correct approach is to use `countplot`.

## Distinction between Usage of sns `countplot` and `barplot`

`countplot` pretty much has a singular purpose, which is to display counts in each category by means of bars.

Unlike how `countplot` deals with categorical variables, `barplot` deals with numerical variables instead. It deals with representing the "central tendency" of that numerical variable, along with error bars to show the uncertainty about the mean.

An example of `barplot`'s use case that fits this analysis, is by grouping the songs into categories representing different genres, and have the y-axis depict the duration of that song. So, end result is that we can have a visual representation of the average duration of songs in K-pop, versus Pop, and Electropop, etc. Don't go scrolling - I did not do it here. I'm just describing an example.

## Categorical Axes

An alternative to `barplot`, would be `heatmap`. A heatmap deals with 2 dimensions of categorical variables to form a 2D matrix. Elements in the matrix are populated with values from a numerical variable, as obtained from the intersection of the 2 chosen dimensions. For example, I pick the 2 dimensions as 'Year' and 'Language'. Then for the numerical variable I can select 'Count'. I have actually done this - it's represented as matplotlib's `area` plot somewhere at the bottom of this notebook. I also added the `heatmap` for good measure, though the outcome doesn't really look that nice.

## Numerical Axes

If the two numerical variables have some form of trend, then you can use a scatter plot. You know, like regression plots? That said, the data being represented (as points) on the coordinate plane would be categorical variable. Like if you axes are "Advertising Costs" and "Revenue", then you have categories such as the different companies that you are analysing. In seaborn (and matplotlib), you can definite colour code points depending on which category they belong in.

If the two variables are like normal bivariate, then you can use sns's `jointplot`. It's like a 2D density plot.

## Creating new variables

You can create new numerical variables by use of aggregate functions (group by categories) like count, mean, etc.

The same can be said for categorical variables, as seen through the use of bins (for timespan) in histograms.
